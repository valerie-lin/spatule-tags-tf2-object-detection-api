```python
import xml.etree.cElementTree as ET
import os
import json
```

```python

template_xml = "path/to/template.xml"
```

```python
cwd = os.getcwd()
photos = os.listdir(cwd + "/path/to/photos")
```

```python
images = []
for item in photos:
    if item[-4:] == ".jpg":
        images.append(item)
        
print(len(images))
```

```python

for image in images:
    image_name = image[:-4]
    annotation = ET.parse(template_xml).getroot()
    annotation.find("path").text += "/" + image_name + ".jpg"
    annotation.find("filename").text = image_name + ".jpg"

    tree = ET.ElementTree()
    tree._setroot(annotation)
    tree.write("/path/to/image/folders/" + image_name + ".xml", encoding="utf-8")
```
